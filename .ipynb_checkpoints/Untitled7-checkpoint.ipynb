```python
Datadirectory = "Training/" ##training dataset
```

```python
Classes = ["angry","disgusted","fearful","happy","neutral","sad","surprised"] # list of classes
```

```python
for category in Classes : 
    path = os.path.join(Datadirectory , category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img))
        plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
        plt.show()
        break
    break
```

```python
img_size = 224
new_array= cv2.resize(img_array,(img_size,img_size))
plt.imshow(cv2.cvtColor(new_array, cv2.COLOR_BGR2RGB))
```

```python
new_array.shape
```

```python
#read all images and covert them to array
training_Data = []
def create_training_data():
    for category in Classes : 
        path = os.path.join(Datadirectory , category)
        class_num = Classes.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array= cv2.resize(img_array,(img_size,img_size))
                training_Data.append([new_array,class_num])
            except Exception as e:
                pass
            
```

```python
create_training_data()
```

```python
print(len(training_Data))
```

```python
import random
random.shuffle(training_Data)
```

```python
X = []
y = []
for features , label in training_Data:
    X.append(features)
    y.append(label)
    
X = np.array(X).reshape(-1,img_size,img_size,3)
```

```python
X.shape
```

```python
X = X/250.0
```

```python
Y = np.array(y)
```

```python
Y.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNet()
```

```python
model.summary()
```

```python
#Transfer Learning - Tuning , weights will start from last check point
```

```python
base_input = model.layers[0].input # input
```

```python
base_output = model.layers[-6].output
```

```python
base_output
```

```python
final_output = layers.Dense(128)(base_output) ## adding new layer , after the output of global pooling layer
final_output = layers.Activation('relu')(final_output) ## activation func
final_output = layers.Dense(64)(final_output)
final_output = layers.Activation('relu')(final_output)
final_output = layers.Dense(7,activation='softmax')(final_output) # 7 classes
```

```python
final_output # output
```

```python
new_model = keras.Model(inputs = base_input, outputs = final_output)
```
